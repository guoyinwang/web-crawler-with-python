```python
import pymongo
import charts
import datetime
```

```python
client = pymongo.MongoClient('localhost',27017)
test = client['test']
item_info = test['item_info']
```

```python
for item in item_info.find().limit(10):
    print(item)
```

```python
pipeline = [
    {'$match':{'time':3}},
    {'$group':{'_id':{'$slice':['$area',1]},'counts':{'$sum':1}}},
    {'$sort' :{'counts':-1}},
    
]
```

```python
for i in item_info.aggregate(pipeline):
    print(i)
```

```python
def data_gen():
    pipeline = [
    {'$match':{'time':3}},
    {'$group':{'_id':{'$slice':['$area',1]},'counts':{'$sum':1}}},
    {'$sort' :{'counts':-1}},
    ]
    
    for i in item_info.aggregate(pipeline):
        yield i['_id'][0], i['counts']
```

```python
options = {
    'chart'   : {'zoomType':'xy'},
    'title'   : {'text': 'pir plot'},
    'subtitle': {'text': 'city zone'},
    }


series =  [{
    'type': 'pie',
    'name': 'pie charts',
    'data':[i for i in data_gen()]

        }]
charts.plot(series,options=options,show='inline')
```
